```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
dataset = load_iris()

X = dataset['data']
y = dataset['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
padosi = np.arange(1,9)
train_accuracy = np.empty(len(padosi))
test_accuracy = np.empty(len(padosi))

for i,k in enumerate(padosi):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    train_accuracy[i] = knn.score(X_train, y_train)
    test_accuracy[i] = knn.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt
plt.title("kNN: Varying neighbours")
plt.plot(padosi, train_accuracy, label = "Train accuracy")
plt.plot(padosi, test_accuracy, label = "Test accuracy")
plt.legend()
plt.xlabel('Neighbours')
plt.ylabel('Accuracy')
plt.show()
```
